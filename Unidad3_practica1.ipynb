```python
#importar Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
#Carga el Dataset
from google.colab import drive
drive.mount('/content/drive')
#ID del archivo
#https:https://drive.google.com/file/d/10jb6-kPt9b-03nwtI_ll8CQeruX1LmZj/view?usp=sharing
file_id = '10jb6-kPt9b-03nwtI_ll8CQeruX1LmZj'
url =  f"https://drive.google.com/uc?id={file_id}"

#cargas Datos
df = pd.read_csv(url)
df.head()
```

#Seleccion de Variables predictoras (x) y Variables a predecir (y)

```python
# Limpiar nombres de columnas
df.columns = df.columns.str.strip()

# Selección de variables
X = df[['height', 'width', 'mass', 'color_score']]
y = df['fruit_name']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
clasificador_KNN = KNeighborsClassifier(n_neighbors=1)
clasificador_KNN.fit(X_train, y_train)

print("Precision entrenamiento:", clasificador_KNN.score(X_train, y_train))
```

```python
y_pred = clasificador_KNN.predict(X_test)

print(classification_report(y_test, y_pred))
print("Precisión:", clasificador_KNN.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(clasificador_KNN, X_test, y_test, n_repeats=30, random_state=42)

importancia = pd.DataFrame({
    "Variable": X.columns,
    "Importancia Media": result.importances_mean,
    "Desviación": result.importances_std
}).sort_values("Importancia Media", ascending=False)

importancia
```

```python
fig4 = px.bar(importancia, x="Variable", y="Importancia Media",
              error_y="Desviación", title="Importancia (KNN)",
              text_auto=".3f", color="Variable")
fig4.show()
```

```python
clasificador_KNN3 = KNeighborsClassifier(n_neighbors=3)
clasificador_KNN3.fit(X_train, y_train)

y_pred = clasificador_KNN3.predict(X_test)

print(classification_report(y_test, y_pred))
print("Precisión:", clasificador_KNN3.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred))
```

```python
clasificador_KNN5 = KNeighborsClassifier(n_neighbors=5)
clasificador_KNN5.fit(X_train, y_train)

y_pred = clasificador_KNN5.predict(X_test)

print(classification_report(y_test, y_pred))
print("Precisión:", clasificador_KNN5.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred))
```

```python
for k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    print(f"k={k} → Precisión: {knn.score(X_test, y_test):.2f}")
```

```python
from sklearn.preprocessing import StandardScaler


X = df[["height", "width", "mass", "color_score"]]
y = df["fruit_name"]

# División con stratify
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, stratify=y
)

# Escalado
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(X_train_scaled, y_train)

y_pred_scaled = knn_scaled.predict(X_test_scaled)

print(classification_report(y_test, y_pred_scaled))
print(confusion_matrix(y_test, y_pred_scaled))
```

```python
result = permutation_importance(knn_scaled, X_test_scaled, y_test, n_repeats=30, random_state=42)

importancia = pd.DataFrame({
    "Variable": X.columns,
    "Importancia Media": result.importances_mean,
    "Desviación": result.importances_std
}).sort_values("Importancia Media", ascending=False)

importancia
```

```python
fig4 = px.bar(importancia, x="Variable", y="Importancia Media",
              error_y="Desviación", title="Importancia escalada",
              text_auto=".3f", color="Variable")
fig4.show()
```

```python
fruta1 = [[5.5, 2.2, 150, 0.70]]
print(knn_scaled.predict(fruta1))

frutas_nuevas = [[5.5, 2.2, 120, 0.70], [2, 2, 160, 0.50]]
print(knn_scaled.predict(frutas_nuevas))
```
